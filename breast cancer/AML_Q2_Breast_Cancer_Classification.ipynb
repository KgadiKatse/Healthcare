```python
# function for fitting trees of various depths on the training data using cross-validation
def run_cross_validation_on_trees(X, y, tree_depths, cv=5, scoring='accuracy'):
    cv_scores_list = []
    cv_scores_std = []
    cv_scores_mean = []
    accuracy_scores = []
    for depth in tree_depths:
        tree_model = DecisionTreeClassifier(criterion='entropy', max_depth=depth)
        cv_scores = cross_val_score(tree_model, X, y, cv=cv, scoring=scoring)
        cv_scores_list.append(cv_scores)
        cv_scores_mean.append(cv_scores.mean())
        cv_scores_std.append(cv_scores.std())
        accuracy_scores.append(tree_model.fit(X, y).score(X, y))
    cv_scores_mean = np.array(cv_scores_mean)
    cv_scores_std = np.array(cv_scores_std)
    accuracy_scores = np.array(accuracy_scores)
    return cv_scores_mean, cv_scores_std, accuracy_scores

# function for plotting cross-validation results
def plot_cross_validation_on_trees(depths, cv_scores_mean, cv_scores_std, accuracy_scores, title):
    fig, ax = plt.subplots(1,1, figsize=(15,5))
    ax.plot(depths, cv_scores_mean, '-o', label='mean cross-validation accuracy', alpha=0.9)
    ax.fill_between(depths, cv_scores_mean-2*cv_scores_std, cv_scores_mean+2*cv_scores_std, alpha=0.2)
    ylim = plt.ylim()
    ax.plot(depths, accuracy_scores, '-*', label='train accuracy', alpha=0.9)
    ax.set_title(title, fontsize=16)
    ax.set_xlabel('Tree depth', fontsize=14)
    ax.set_ylabel('Accuracy', fontsize=14)
    ax.set_ylim(ylim)
    ax.set_xticks(depths)
    ax.legend()

# fitting trees of depth 1 to 11
sm_tree_depths = range(1,11)
sm_cv_scores_mean, sm_cv_scores_std, sm_accuracy_scores = run_cross_validation_on_trees(x_train_SMOTE, y_train_SMOTE, sm_tree_depths)

# plotting accuracy
plot_cross_validation_on_trees(sm_tree_depths, sm_cv_scores_mean, sm_cv_scores_std, sm_accuracy_scores,
                               'Accuracy per decision tree depth on training data')
```

```python
idx_max = sm_cv_scores_mean.argmax()
sm_best_tree_depth = sm_tree_depths[idx_max]
sm_best_tree_cv_score = sm_cv_scores_mean[idx_max]
sm_best_tree_cv_score_std = sm_cv_scores_std[idx_max]
print('The depth-{} tree achieves the best mean cross-validation accuracy {} +/- {}% on training dataset'.format(
      sm_best_tree_depth, round(sm_best_tree_cv_score*100,5), round(sm_best_tree_cv_score_std*100, 5)))
```

```python
# function for training and evaluating a tree
def run_single_tree(X_train, y_train, X_test, y_test, depth):
    model = DecisionTreeClassifier(criterion='entropy', max_depth=depth).fit(X_train, y_train)
    accuracy_train = model.score(X_train, y_train)
    accuracy_test = model.score(X_test, y_test)
    print('Single tree depth: ', depth)
    print('Accuracy, Training Set: ', round(accuracy_train*100,5), '%')
    print('Accuracy, Test Set: ', round(accuracy_test*100,5), '%')
    return accuracy_train, accuracy_test


# train and evaluate a 5-depth tree
sm_best_tree_accuracy_train, sm_best_tree_accuracy_test = run_single_tree(x_train_SMOTE, y_train_SMOTE,
                                                                          x_test_SMOTE, y_test_SMOTE,
                                                                          sm_best_tree_depth)
```

```python
# KNN is partly robust to missing values but it is best for them to be imputed for the sake of implementation.

train_data2 = bc_train_data1.copy()
test_data2 = bc_test_data1.copy()

training_missing_columns = train_data2[train_features]
#Technique 1: Using median to impute the missing values in training set
for i in training_missing_columns:
 train_data2.loc[train_data2.loc[:,i].isnull(),i]=train_data2.loc[:,i].median()

testing_missing_columns = test_data2[test_features]
#Technique 1: Using median to impute the missing values in test set
for i in training_missing_columns:
 test_data2.loc[test_data2.loc[:,i].isnull(),i]=test_data2.loc[:,i].median()
```

```python
# split train and test descriptive feature and target feature

# make a copy of the datasets for training and testing
k_train_data = train_data2.copy()
k_test_data = test_data2.copy()


# set up the separate train and test set descriptive features and target feature
train_xk = k_train_data[train_features]
test_xk = k_test_data[test_features]

train_yk = k_train_data['diagnosis']
test_yk = k_test_data['diagnosis']
```

```python
# Normalize the data for knn implementation

# returns a numpy array of the dataframe
x_train_k = train_xk.values
x_test_k = test_xk.values

min_max_scaler = preprocessing.MinMaxScaler()

# scale features
x_train_knorm = min_max_scaler.fit_transform(x_train_k)
x_test_knorm = min_max_scaler.fit_transform(x_test_k)

# return to dataframe form
x_train_knn = pd.DataFrame(x_train_knorm)
x_test_knn = pd.DataFrame(x_test_knorm)
```

Dealing with the imbalanced data

SMOTE

```python
# TRAINING
# summarize class distribution
print("Before oversampling: ",Counter(train_yk))

# define oversampling strategy
oversampling = SMOTE(random_state=42)

# fit and apply the transform
x_train_SMOTE1, y_train_SMOTE1 = oversampling.fit_resample(x_train_knn, train_yk)

# summarize class distribution
print("After oversampling: ",Counter(y_train_SMOTE1))

#################

# TESTING
# summarize class distribution
print("\nBefore oversampling: ",Counter(test_yk))

# fit and apply the transform
x_test_SMOTE1, y_test_SMOTE1 = oversampling.fit_resample(x_test_knn, test_yk)

# summarize class distribution
print("After oversampling: ",Counter(y_test_SMOTE1))
```

TOMEK LINKS

```python
# TRAINING
# summarize class distribution
print("Training Data")
print("Before undersampling: ",Counter(train_yk))

# define the undersampling method
undersample = TomekLinks()

# fit and apply the transform
x_train_TL1, y_train_TL1 = undersample.fit_resample(x_train_knn, train_yk)

# summarize class distribution
print("After undersampling: ",Counter(y_train_TL1))

#################

# TESTING
# summarize class distribution
print("\n Testing Data")
print("Before undersampling: ",Counter(test_yk))

# fit and apply the transform
x_test_TL1, y_test_TL1 = undersample.fit_resample(x_test_knn, test_yk)

# summarize class distribution
print("After undersampling: ",Counter(y_test_TL1))
```

KNN Implemntation - when imbalanced data is oversampled

```python
neighbors = np.arange(1, 60)
knn_train_accuracy = np.empty(len(neighbors))
knn_test_accuracy = np.empty(len(neighbors))

# Loop over K values
for i, k in enumerate(neighbors):
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(x_train_SMOTE1, y_train_SMOTE1)

    # Compute training and test data accuracy
    knn_train_accuracy[i] = knn.score(x_train_SMOTE1, y_train_SMOTE1)
    knn_test_accuracy[i] = knn.score(x_test_SMOTE1, y_test_SMOTE1)

# Generate plot
plt.plot(neighbors, knn_test_accuracy, label = 'Testing dataset Accuracy')
plt.plot(neighbors, knn_train_accuracy, label = 'Training dataset Accuracy')

plt.legend()
plt.xlabel('Number of neighbors')
plt.ylabel('Accuracy')
plt.show()
```

```python
# optimum value of k = 25

knn_25 = KNeighborsClassifier(n_neighbors=25)

knn_25.fit(x_train_SMOTE1, y_train_SMOTE1)

# Predict on dataset which model has not seen before
knn_predict0 = knn_25.predict(x_test_SMOTE1)

print('Confusion Matrix:')
print(confusion_matrix(y_test_SMOTE1, knn_predict0))
print('\nClassification Report')
print(classification_report(y_test_SMOTE1, knn_predict0))
print("\nAccuracy Score on test set:")
print(knn_25.score(x_test_SMOTE1, y_test_SMOTE1))
```
